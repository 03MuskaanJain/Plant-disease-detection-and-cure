```python
import os.path

from keras.models import load_model
import tkinter as tk
from tkinter import filedialog
import pydicom as dicom
import pneumonia_detection_project_functions as util
import cv2
from matplotlib import pyplot as plt
import matplotlib.patches as patches
import numpy as np
```

```python
segmentation_model = load_model('resnet_segmentation_model.h5', compile=False)
```

```python
pneumonia_dataframe = util.get_pneumonia_dataframe()
print('column data type \n' + str(pneumonia_dataframe.dtypes))
```

```python
root = tk.Tk()
root.withdraw()

file_path = filedialog.askopenfilename()
# file_path = './rsna-pneumonia-detection-challenge/stage_2_train_images/00436515-870c-4b36-a041-de91049b9ab4.dcm'
image = dicom.dcmread(file_path)

resized_image = util.resize_and_reshape([image.pixel_array])
prediction = segmentation_model.predict(resized_image)[0]
# resizing predicted mask
prediction = cv2.resize(prediction, (1024, 1024)).reshape(1024, 1024, 1)


file_id = os.path.basename(file_path).split('.')[0]
mask_bbox = pneumonia_dataframe[pneumonia_dataframe['patient_id'] == file_id]['bbox_data'].tolist()[0]
print(file_id, mask_bbox )


plt.figure(figsize=(10, 10))
ax1 = plt.subplot(1, 1, 1)
ax1.imshow(image.pixel_array, cmap='gray', label='Original Pneumonia Segments')
mask = np.expand_dims(util.get_segmentation_mask(image.pixel_array, bounding_box_data=mask_bbox), -1)
segments = util.get_image_segments(mask, 0)
for region in segments:
    (x, y, width, height) = region
    ax1.add_patch(patches.Rectangle((x, y), width, height, linewidth=2, edgecolor='b', facecolor='none'))
plt.title('Original Pneumonia Segments')
plt.show()


plt.figure(figsize=(10, 10))
ax1 = plt.subplot(1, 1, 1)
ax1.imshow(image.pixel_array, cmap='gray')
segments = util.get_image_segments(prediction, 0.5)
for region in segments:
    (x, y, width, height) = region
    ax1.add_patch(patches.Rectangle((x, y), width, height, linewidth=2, edgecolor='g', facecolor='none'))
plt.title('Predicted Pneumonia Segments')
plt.show()
```
