```python
def iou_loss(y_true, y_pred):
    y_true = tf.reshape(y_true, [-1])
    y_pred = tf.reshape(y_pred, [-1])
    intersection = tf.reduce_sum(y_true * y_pred)
    score = (intersection + 1.) / (tf.reduce_sum(y_true) + tf.reduce_sum(y_pred) - intersection + 1.)
    return 1 - score


def iou_bce_loss(y_true, y_pred):
    return 0.5 * keras.losses.binary_crossentropy(y_true, y_pred) + 0.5 * iou_loss(y_true, y_pred)


def mean_iou(y_true, y_pred):
    y_pred = tf.round(y_pred)
    intersect = tf.reduce_sum(y_true * y_pred, axis=[1, 2, 3])
    union = tf.reduce_sum(y_true, axis=[1, 2, 3]) + tf.reduce_sum(y_pred, axis=[1, 2, 3])
    smooth = tf.ones(tf.shape(intersect))
    return tf.reduce_mean((intersect + smooth) / (union - intersect + smooth))


# cosine learning rate annealing
def cosine_annealing(x):
    lr = 0.001
    epochs = 25
    return lr * (np.cos(np.pi * x / epochs) + 1.) / 2


# def dice_coef(y_true, y_pred):
#     from keras.backend import flatten, sum
#     y_true_f = flatten(y_true)
#     y_pred_f = flatten(y_pred)
#     intersection = sum(y_true_f * y_pred_f)
#     return (2. * intersection + 1) / (sum(y_true_f) + sum(y_pred_f) + 1)


def dice_coef(y_true, y_pred):
    from keras.backend import flatten, sum
    y_true_f = flatten(y_true)
    y_pred_f = flatten(y_pred)
    intersection = sum(y_true_f * y_pred_f)
    return (2. * intersection + 0.0001) / (sum(y_true_f) + sum(y_pred_f) + 0.0001)


def dice_coef_loss(y_true, y_pred):
    return 1-dice_coef(y_true, y_pred)
```

```python
import keras

def down_sample_blocks(channels, inputs):
    x = BatchNormalization(momentum=0.9)(inputs)
    x = LeakyReLU(0)(x)
    x = Conv2D(channels, 1, padding='same', use_bias=False)(x)
    x = MaxPool2D(2)(x)
    return x


def residual_blocks(channels, inputs):
    x = BatchNormalization(momentum=0.9)(inputs)
    x = LeakyReLU(0)(x)
    x = Conv2D(channels, 3, padding='same', use_bias=False)(x)
    x = BatchNormalization(momentum=0.9)(x)
    x = LeakyReLU(0)(x)
    x = Conv2D(channels, 3, padding='same', use_bias=False)(x)
    return add([x, inputs])


def create_resnet(input_size, channels, block_size=2, depth=5):
    # input
    inputs = Input(shape=input_size)
    # resized_inputs = Lambda(lambda image: tf.image.resize(image, target_size))(inputs)
    x = Conv2D(channels, 3, padding='same', use_bias=False)(inputs)
    # residual blocks
    for d in range(depth):
        channels = channels * 2
        x = down_sample_blocks(channels, x)
        for b in range(block_size):
            x = residual_blocks(channels, x)
    # output
    x = BatchNormalization(momentum=0.9)(x)
    x = LeakyReLU(0)(x)
    x = Conv2D(1, 1, activation='sigmoid')(x)
    outputs = UpSampling2D(2 ** depth)(x)
    return keras.Model(inputs=inputs, outputs=outputs)
```

```python
def create_unet(input_size):
    inputs = Input(input_size)

    conv1 = Conv2D(32, (3, 3), activation='relu', padding='same')(inputs)
    conv1 = Conv2D(32, (3, 3), activation='relu', padding='same')(conv1)
    pool1 = MaxPooling2D(pool_size=(2, 2))(conv1)

    conv2 = Conv2D(64, (3, 3), activation='relu', padding='same')(pool1)
    conv2 = Conv2D(64, (3, 3), activation='relu', padding='same')(conv2)
    pool2 = MaxPooling2D(pool_size=(2, 2))(conv2)

    conv3 = Conv2D(128, (3, 3), activation='relu', padding='same')(pool2)
    conv3 = Conv2D(128, (3, 3), activation='relu', padding='same')(conv3)
    pool3 = MaxPooling2D(pool_size=(2, 2))(conv3)

    conv4 = Conv2D(256, (3, 3), activation='relu', padding='same')(pool3)
    conv4 = Conv2D(256, (3, 3), activation='relu', padding='same')(conv4)
    pool4 = MaxPooling2D(pool_size=(2, 2))(conv4)

    conv5 = Conv2D(512, (3, 3), activation='relu', padding='same')(pool4)
    conv5 = Conv2D(512, (3, 3), activation='relu', padding='same')(conv5)

    up6 = concatenate([Conv2DTranspose(256, (2, 2), strides=(2, 2), padding='same')(conv5), conv4], axis=3)
    conv6 = Conv2D(256, (3, 3), activation='relu', padding='same')(up6)
    conv6 = Conv2D(256, (3, 3), activation='relu', padding='same')(conv6)

    up7 = concatenate([Conv2DTranspose(128, (2, 2), strides=(2, 2), padding='same')(conv6), conv3], axis=3)
    conv7 = Conv2D(128, (3, 3), activation='relu', padding='same')(up7)
    conv7 = Conv2D(128, (3, 3), activation='relu', padding='same')(conv7)

    up8 = concatenate([Conv2DTranspose(64, (2, 2), strides=(2, 2), padding='same')(conv7), conv2], axis=3)
    conv8 = Conv2D(64, (3, 3), activation='relu', padding='same')(up8)
    conv8 = Conv2D(64, (3, 3), activation='relu', padding='same')(conv8)

    up9 = concatenate([Conv2DTranspose(32, (2, 2), strides=(2, 2), padding='same')(conv8), conv1], axis=3)
    conv9 = Conv2D(32, (3, 3), activation='relu', padding='same')(up9)
    conv9 = Conv2D(32, (3, 3), activation='relu', padding='same')(conv9)

    conv10 = Conv2D(1, (1, 1), activation='sigmoid')(conv9)

    return keras.models.Model(inputs=[inputs], outputs=[conv10])
```

```python
def train_model(model, train_data, validation_data, chart_var, callbacks=None, epochs=25, batch_size=32):
    from keras_preprocessing.image import ImageDataGenerator

    (test_x, test_y) = train_data
    (val_x, val_y) = validation_data
    image_generator_with_augmentation = ImageDataGenerator(rotation_range=30, zoom_range=0.20,
                                                           fill_mode="nearest", shear_range=0.20, horizontal_flip=True,
                                                           width_shift_range=0.1, height_shift_range=0.1)
    history = model.fit_generator(image_generator_with_augmentation.flow(x=test_x, y=test_y, batch_size=batch_size),
                                  validation_data=(val_x, val_y), callbacks=callbacks,
                                  steps_per_epoch=len(X_train) // 32,
                                  epochs=epochs)  # workers=4, use_multiprocessing=True
    plt.figure(figsize=(12, 4))
    for index, var in enumerate(chart_var):
        plt.subplot(1, 3, (index / 2) + 1)
        plt.plot(history.epoch, history.history[var], label=var)
        plt.legend()
    plt.show()
```

```python
from keras_preprocessing.image import ImageDataGenerator


def print_image_prediction(model, x_val, y_val, row=5, col=5, threshold=0.5):
    import matplotlib.patches as patches

    for image_batch, mask_batch in ImageDataGenerator().flow(x_val, y_val, batch_size=row * col):
        predictions = model.predict(image_batch)
        fig, axes = plt.subplots(row, col, figsize=(20, 15))
        axes = axes.ravel()
        axes_idx = 0
        for image, mask, prediction in zip(image_batch, mask_batch, predictions):
            # add image to the plot
            axes[axes_idx].imshow(image[:, :, 0], cmap='gray')
            segments = util.get_image_segments(mask, 0)
            for region in segments:
                (x, y, width, height) = region
                axes[axes_idx].add_patch(patches.Rectangle((x, y), width, height, linewidth=2, edgecolor='b', facecolor='none'))
            segments = util.get_image_segments(prediction, 0)
            for region in segments:
                (x, y, width, height) = region
                axes[axes_idx].add_patch(patches.Rectangle((x, y), width, height, linewidth=2, edgecolor='g', facecolor='none'))
            axes_idx += 1
        plt.show()
        # stop after first batch is completed, so that it only prints up to 'number_of_samples' input
        return
```

```python
def write_test_prediction(model, threshold=0.5):
    from cv2 import cv2
    from skimage import measure

    util.section('Predicting all test images')

    data_path = util.get_data_paths()
    (_, test_image_dir) = data_path['image_dir']
    data_frame = pd.DataFrame(columns=['patient_id', 'prediction_string'])
    for index, filename in enumerate(os.listdir(test_image_dir)):
        image = dicom.dcmread(os.path.join(test_image_dir, filename))
        pixel_data = util.resize_and_reshape([image.pixel_array], TARGET_HEIGHT, TARGET_WIDTH)
        prediction = model.predict(pixel_data)[0]
        # resizing predicted mask
        prediction = cv2.resize(prediction, (1024, 1024)).reshape(1024, 1024, 1)
        # check predicted pixels confidence is above threshold
        components = prediction[:, :, 0] > threshold
        components = measure.label(components)
        prediction_string = ''
        for region in measure.regionprops(components):
            y, x, y2, x2 = region.bbox
            height = y2 - y
            width = x2 - x
            confidence = np.mean(prediction[y:y + height, x:x + width])
            prediction_string += f"{confidence}\t{x}\t{y}\t{width}\t{height}\t"
        patient_id = filename.split('.')[0]
        data_frame.loc[index] = {'patient_id': patient_id, 'prediction_string': prediction_string}
    return data_frame
```

```python
# GLOBAL VARIABLES
TARGET_HEIGHT = 256
TARGET_WIDTH = 256
CHANNELS = 1
SEED = 42
```

```python
sample_df = util.get_sampled_data_for_modelling(pneumonia_dataframe, number_of_patients_record=3000)

print('Number patient ids in sampled data -', len(sample_df['patient_id'].unique()))
print('Number of row and columns in sampled data -', sample_df.shape)
print("\n\n" + "Sampled Data For Modelling".center(100, "."), sample_df.head(2).T)

# shuffle images
sample_df = sample_df.sample(frac=1, random_state=SEED).reset_index(drop=True)
```

```python
from sklearn.model_selection import train_test_split

X = util.resize_and_reshape(sample_df.pixel_data, TARGET_HEIGHT, TARGET_WIDTH)
Y = util.resize_and_reshape(sample_df.pixel_mask, TARGET_HEIGHT, TARGET_WIDTH, mask=True)
print('Original Image and Mask Image Input Shapes For Modelling', X.shape, Y.shape)
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=SEED)
```

```python
util.section('Sample Image Visualisation')
print_count = 0
for train_img, train_msk in zip(X_train, y_train):
    util.view_dicom_image(train_img, train_msk)
    print(print_count, set([gh for th in train_msk for tt in th for gh in tt]))
    if print_count < 10:
        print_count += 1
        continue
    break

```

```python
model1 = create_resnet(input_size=(TARGET_HEIGHT, TARGET_WIDTH, CHANNELS), channels=32, block_size=2, depth=4)
model1.compile(optimizer='adam', loss=iou_bce_loss, metrics=['accuracy', mean_iou])
model1.summary()

visualkeras.layered_view(model1, legend=True, spacing=5)
```

```python

model_checkpoint = ModelCheckpoint('resnet_segmentation_model.h5', monitor='loss', verbose=1, save_best_only=True)
learning_rate = LearningRateScheduler(cosine_annealing)
early_stopping = EarlyStopping(monitor="val_loss", mode="min", patience=3)

train_model(model1, train_data=(X_train, y_train), validation_data=(X_test, y_test),
            callbacks=[learning_rate, model_checkpoint, early_stopping], epochs=10,
            chart_var=['accuracy', 'val_accuracy', 'loss', 'val_loss', 'mean_iou', 'val_mean_iou'])
```
